````python
import yfinance as yf
import pandas as pd
import plotly.express as px
import numpy as np
````

````python
sp500_hist = sp500.history("30y")
gbpusd_hist = gbpusd.history("30y")
sp500_hist.head()
````

````python
from dataclasses import dataclass
from typing import Dict, List
from enum import Enum


class DataType(Enum):
  STR = 0
  UUID = 1
  INT = 2
  FLOAT = 3



@dataclass
class FieldProfile(object):
  name: str
  dtype: DataType
  unique_values: int
  count: int
  # data profile for types


@dataclass
class Association(object):
  t1: str  # cardinality??
  t2: str
  t1_key: str
  t2_key: str


@dataclass
class Entity(object):
  name: str
  cols: Dict[str, FieldProfile]
  business_key: str
  est_business_keys: int 
  est_records: int
  links: Dict[str, str]




  def __init__(self) -> None:
    pass

  @classmethod
  def build():
    pass


@dataclass
class Model(object):
  name: str
  relations: Dict[str, Entity]



````

````python
def trendify(df: pd.DataFrame)-> pd.DataFrame:
  df["closing_200ma"] = df["Close"].rolling(200).mean()
  df["closing_100ma"] = df["Close"].rolling(100).mean()
  df["closing_50ma"] = df["Close"].rolling(50).mean()
  df["closing_20ma"] = df["Close"].rolling(20).mean()
  df["dt"] = df.index.date
  return df

````

````python
sp500_hist = trendify(sp500_hist)
gbpusd_hist = trendify(gbpusd_hist)
````

````python
px.line(sp500_hist, x="dt", y=["Close", "closing_200ma","closing_100ma", "closing_50ma", "closing_20ma"])
````
